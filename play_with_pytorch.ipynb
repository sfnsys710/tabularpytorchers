```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
del a, b, c

a = torch.tensor(10., requires_grad=True)
b = torch.tensor(20., requires_grad=True)

c = a * b

print()
```

```python
a = torch.tensor([1., 2.], requires_grad=True)
b = torch.tensor([3., 4.], requires_grad=True)

c = a * b

c.backward()

print(a.grad)
print(b.grad)
```
